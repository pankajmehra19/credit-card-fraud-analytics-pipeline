```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv(
    r"D:\Credit-card-analytics-pipeline\data\processed\clean_transactions.csv"
)
```

```python
engine = create_engine(
    'postgresql://postgres:123456789@localhost:5432/credit_card_pipeline'
    )
```

```python
df['transaction_date'] = pd.to_datetime(
    df['transaction_date']
)
```

```python
df['transaction_datetime'] = pd.to_datetime(
    df['transaction_datetime']
)
```

```python
engine
```

```python
df.to_sql(
    name = 'transactions',
    con = engine,
    if_exists = 'replace',
    index = False
)
```
