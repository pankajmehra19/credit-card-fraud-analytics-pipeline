```python
df = pd.read_csv(r"D:\Credit-card-analytics-pipeline\data\processed\transactions_sample.csv")
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('?', '', regex = False)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['merchant_state'] = df['merchant_state'].fillna('unknown')
```

```python
df['zip'] = df['zip'].fillna(0)
```

```python
df['errors'] = df['errors'].fillna('No error')
```

```python
df.info()
```

```python
df['amount'] = df['amount'].str.replace('$', '', regex = False)
```

```python
df['amount'] = df['amount'].astype(float)
```

```python
df.head()
```

```python
df['transaction_date'] = pd.to_datetime(
    df[['year','month','day']])
```

```python
df['transaction_datetime'] = pd.to_datetime(
    df['transaction_date'].astype(str) + ' ' + df['time'])
```

```python
df[['transaction_date', 'transaction_datetime']].head()
```

```python
df['transaction_datetime'].dtype
```

```python
df['is_fraud'].value_counts()
```

```python
df['is_fraud_flag'] = df['is_fraud'].map({
    'Yes': 1,
    'No':0
})
```

```python
df['errors'].value_counts()
```

```python
df['error_flag'] = df['errors'].apply(
    lambda x: 0 if x == 'No Error' else 1
)
```

```python
df[['is_fraud', 'is_fraud_flag', 'errors', 'error_flag']].head()
```

```python
df.info()
```

```python
df = df.drop(
    columns=['year', 'month', 'day', 'time', 'is_fraud']
)
```

```python
df.columns
```

```python
#Total Transaction Amount
df['amount'].sum()
```

```python
#Average Transaction Amount
df['amount'].mean()
```

```python
#Fraud Rate
df['is_fraud_flag'].mean()*100
```

```python
df['use_chip'].value_counts()
```

```python
df.groupby('use_chip')['is_fraud_flag'].mean() * 100
```

```python
df.groupby('merchant_city')['is_fraud_flag'].mean().sort_values(ascending=False).head(10)*100
```

```python
city_stats = df.groupby('merchant_city').agg(
    total_transactions = ('is_fraud_flag', 'count'),
    fraud_rate=('is_fraud_flag', 'mean')
)
```

```python
city_stats.head()
```

```python
filtered_cities = city_stats[
    city_stats['total_transactions'] >= 100
]
```

```python
filtered_cities.sort_values(
    by = 'fraud_rate',
    ascending = False
).head(10)
```

```python
mcc_stats = df.groupby('mcc').agg(
    total_transactions = ('is_fraud_flag', 'count'),
    fraud_rate = ('is_fraud_flag', 'mean')
)
```

```python
mcc_stats.sort_values(
    by = 'fraud_rate',
    ascending = False
).head(10)
```

```python
filtered_mcc = mcc_stats[
    mcc_stats['total_transactions'] >= 100
]
```

```python
filtered_mcc.sort_values(
    by = 'fraud_rate',
    ascending = False
).head(10)
```

```python
df['transaction_hour'] = df['transaction_datetime'].dt.hour
```

```python
hourly_stats = df.groupby('transaction_hour').agg(
    total_transactions = ('amount', 'count'),
    fraud_rate = ('is_fraud_flag', 'mean')
)
```

```python
hourly_stats
```

```python
daily_stats = df.groupby('transaction_date').agg(
    total_transactions= ('amount', 'count'),
    total_amount = ('amount', 'sum'),
    fraud_rate = ('is_fraud_flag', 'mean')

)
```

```python
daily_stats.head()
```
